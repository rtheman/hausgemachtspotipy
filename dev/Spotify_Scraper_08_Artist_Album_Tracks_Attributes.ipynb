# 3.) Albums

## a. Populate Artist's Albums

List all albums of an artist via artist's Spotify ID

```python
row_albums = []

for index, row_artist in df_artists.iterrows():
    
    albums = sp.get_artist_albums(row_artist['artist_id'])

    # --- #HardCode "20" for max. number of rows returned by Spotify 
    if albums["total"] > 20:
        albums_cnt = 20
    else:
        albums_cnt = albums["total"]
    
    for i in range(albums_cnt):
        artist_id = row_artist['artist_id']
        album_id = albums["items"][i]["id"]
        album_release_date = albums["items"][i]["release_date"]
        album_name = albums["items"][i]["name"]
        row_albums.append([artist_id, album_id, album_release_date, album_name])



df_albums = pd.DataFrame(row_albums, columns=["artist_id", "album_id", "album_release_date", "album_name"])
df_albums.to_clipboard(sep = ",", index=False)

compression_opts = dict(method='zip', archive_name='out_albums.csv')  
df_albums.to_csv('./results/out_albums.zip', index=False, compression=compression_opts)  
```

---

# 4.) Tracks

## a. List Tracks per Album

LIst all tracks of an album via album's Spotify ID

```python
row_tracks = []

for index, row_album in df_albums.iterrows():

    tracks = sp.get_album_tracks(row_album['album_id'])

    # --- #HardCode "20" for max. number of rows returned by Spotify 
    if tracks["total"] > 20:
        track_cnt = 20
    else:
        track_cnt = tracks["total"]

    for i in range(track_cnt):
        album_id = row_album['album_id']
        track_id = tracks["items"][i]["id"]
        track_num = tracks["items"][i]["track_number"]
        track_name = tracks["items"][i]["name"]
        # track_duration = tracks["items"][i]["duration_ms"]
        track_duration = datetime.timedelta(milliseconds = tracks["items"][i]["duration_ms"])
        row_tracks.append([album_id, track_id, track_num, track_name, track_duration])


df_tracks = pd.DataFrame(row_tracks, columns=["album_id", "track_id", "track_num", "track_name", "track_duration"])
df_tracks.to_clipboard(sep = ",", index=False)

# compression_opts = dict(method='zip', archive_name='out_tracks.csv')  
# df_tracks.to_csv('./results/out_tracks.zip', index=False, compression=compression_opts)  
```

## b. Query track's Audio Features

```python
danceability = []
energy = []
key = []
loudness = []
speechiness = []
acousticness = []
instrumentalness = []
liveness = []
valence = []
tempo = []

for id in df_tracks["track_id"]:
    track_resp = sp.get_track_features(id)
    danceability.append( track_resp["danceability"] )
    energy.append( track_resp["energy"] )
    key.append( track_resp["key"] )
    loudness.append( track_resp["loudness"] )
    speechiness.append( track_resp["speechiness"] )
    acousticness.append( track_resp["acousticness"] )
    instrumentalness.append( track_resp["instrumentalness"] )
    liveness.append( track_resp["liveness"] )
    valence.append( track_resp["valence"] )
    tempo.append( track_resp["tempo"] )

df_tracks["danceability"] = danceability
df_tracks["energy"] = energy
df_tracks["key"] = key
df_tracks["loudness"] = loudness
df_tracks["speechiness"] = speechiness
df_tracks["acousticness"] = acousticness
df_tracks["instrumentalness"] = instrumentalness
df_tracks["liveness"] = liveness
df_tracks["valence"] = valence
df_tracks["tempo"] = tempo

compression_opts = dict(method='zip', archive_name='out_tracks.csv')  
df_tracks.to_csv('./results/out_tracks.zip', index=False, compression=compression_opts)  
```
